Linear regression

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# simulate data for linear relationship between pageSpeeds and purchaseAmount
pageSpeeds_linear = np.random.normal(3, 1, 1000)
purchaseAmount_linear = 100 - (pageSpeeds_linear + np.random.normal(0, 0.1, 1000)) * 3

plt.scatter(pageSpeeds_linear, purchaseAmount_linear)
```

```python
from scipy import stats
slope_l, intercept_l, r_value_l, p_value_l, std_err_l = stats.linregress(pageSpeeds_linear, purchaseAmount_linear)
```

```python
r_value_l ** 2
```

```python
def predict(x):
    return slope_l * x + intercept_l

fitLine = predict(pageSpeeds_linear)

plt.scatter(pageSpeeds_linear, purchaseAmount_linear)
plt.plot(pageSpeeds_linear, fitLine, c="r")
plt.show()
```

Polynomial regression

```python
np.random.seed(2)
pageSpeeds_poly = np.random.normal(3, 1, 1000)
purchaseAmount_poly = np.random.normal(50, 10, 1000) / pageSpeeds_poly
plt.scatter(pageSpeeds_poly, purchaseAmount_poly)
```

```python
p4 = np.poly1d(np.polyfit(pageSpeeds_poly, purchaseAmount_poly, 4))
```

```python
p4
```

```python
0.54005597 * (3 ** 4) + (-8.85641318) * (3 ** 3) + 52.25378374 * (3 ** 2) + (-135.34422815) * 3 + 147.6050662
```

```python
p4([1,2,3])
```

```python
np.linspace(0, 7, 100)
```

```python
xp = np.linspace(0, 7, 100)
plt.scatter(pageSpeeds_poly, purchaseAmount_poly)
plt.plot(xp, p4(xp), c="r")
plt.show()
```

```python
from sklearn.metrics import r2_score

r2_poly = r2_score(purchaseAmount_poly, p4(pageSpeeds_poly))
```

Multiple regression

```python
import pandas as pd
```

```python
car_df = pd.read_excel("./data/regression/cars.xls")
```

```python
car_df
```

```python
df1=car_df[["Mileage", "Price"]]
bins=np.arange(0, 50000, 10000)
pd.cut(df1["Mileage"], bins)
```

```python
groups = df1.groupby(pd.cut(df1["Mileage"], bins)).mean()
```

```python
groups["Price"].plot.line()
```

```python
import statsmodels.api as sm
from sklearn.preprocessing import StandardScaler
scale = StandardScaler()

X = car_df[["Mileage", "Cylinder", "Doors"]]
y = car_df["Price"]

X[["Mileage", "Cylinder", "Doors"]] = scale.fit_transform(X[["Mileage", "Cylinder", "Doors"]])

print(X)

est = sm.OLS(y, X).fit()

est.summary()
```

```python
y.groupby(car_df.Doors).mean()
```

```python
# prediction of car price if mileage is 45000, cylinder=8, and doors=4
scaled = scale.transform([[45000, 8, 4]])
print(scaled)
predicted = est.predict(scaled)
print(predicted)
```
