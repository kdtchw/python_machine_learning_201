```python
# mnist database contains handwritten digit image data
from tensorflow.keras.datasets import mnist 
```

```python
# digit ranges from 0 to 9
num_classes = 10

# each digit image is 28 * 28 pixels, represented by 2d array 28 rows * 28 columns
num_features = 784

(X_train, y_train), (X_test, y_test) = mnist.load_data()
```

```python
# each pixel in digit image ranges from 0 to 255 (rgb value) (black and white image)
X_train[0]
```

```python
# convert to float 32
X_train, X_test = np.array(X_train, np.float32), np.array(X_test, np.float32)
```

```python
X_train[0]
```

```python
# convert 2d array to 1d array
X_train = X_train.reshape([-1, num_features])
X_test = X_test.reshape([-1, num_features])
```

```python
# convert each value from [0, 255] to [0, 1]
X_train, X_test = X_train / 255, X_test / 255
```

```python
# visualize the original 2D-array digit image data
def display_sample(num):
    #Print this sample's label 
    label = y_train[num] 
    
    #Reshape the 784 values to a 28x28 image
    image = X_train[num].reshape([28,28])
    plt.title("Sample: %d  Label: %d" % (num, label))
    plt.imshow(image, cmap=plt.get_cmap("gray_r"))
    plt.show()
    
display_sample(1000)
```

```python
learning_rate = 0.001
training_steps = 3000
batch_size = 250
display_step = 100

n_hidden = 512
```

```python
train_data = tf.data.Dataset.from_tensor_slices((X_train, y_train))
train_data
```

```python
train_data.repeat().shuffle(60000).batch(batch_size)
```

```python
# tensorflow repeat, shuffle, and batch, and end by prefetching the 1st element (i.e. 1st batch in this case)
train_data = train_data.repeat().shuffle(60000).batch(batch_size).prefetch(1)
```

```python
train_data
```

```python
# 2-d array variable for storing weights
# 1-d array variable for storing bias

# num_features = 784 (28 * 28 pixels), num_hidden = 512 (number of neurons), num_classes = 10 (10 digits)

# create initializer following normal distribution and apply it to the weights matrix
random_normal = tf.initializers.RandomNormal()

weights = {
    "h": tf.Variable(random_normal([num_features, n_hidden])),
    "out": tf.Variable(random_normal([n_hidden, num_classes]))
}
biases = {
    "b": tf.Variable(tf.zeros([n_hidden])),
    "out": tf.Variable(tf.zeros([num_classes]))
}
```

```python
def neural_net(inputData):
    
    # multiply (1 x 784 matrix) inputData by 2-d array (784 x 512) variable weights["h"] (weights of hidden_neurons),
    # then add the 1-d array bias variable to the product (1 x 512 matrix)
    hidden_layer = tf.add(tf.matmul(inputData, weights["h"]), biases["b"])
    
    # apply activation function (sigmoid) to hidden_layer output for non-linearity.
    hidden_layer = tf.nn.sigmoid(hidden_layer)
    
    # multiply hidden_layer (1 x 512 matrix) by the weights of output neurons (512 x 10 matrix)
    # then add bias of output neurons to the product (1 x 10 matrix)
    out_layer = tf.matmul(hidden_layer, weights["out"]) + biases["out"]
    # apply softmax to the output layer neurons
    return tf.nn.softmax(out_layer)
```

```python
def cross_entropy(y_pred, y_true):
    # apply one hot encoding
    y_true = tf.one_hot(y_true, depth=num_classes)
    # clip prediction values (min: 1e-9, max: 1) to avoid log(0) error
    y_pred = tf.clip_by_value(y_pred, 1e-9, 1.)
    # compute cross-entropy.
    return tf.reduce_mean(-tf.reduce_sum(y_true * tf.math.log(y_pred)))
```

```python
# stochastic gradient descent
optimizer = tf.keras.optimizers.SGD(learning_rate)

def run_optimization(x, y):
    # Wrap computation inside a GradientTape for automatic differentiation.
    with tf.GradientTape() as g:
        pred = neural_net(x)
        loss = cross_entropy(pred, y)
        
    # Variables to update, i.e. trainable variables.
    trainable_variables = list(weights.values()) + list(biases.values())

    # Compute gradients.
    gradients = g.gradient(loss, trainable_variables)
    
    # Update W and b following gradients.
    optimizer.apply_gradients(zip(gradients, trainable_variables))
```

```python
def accuracy(y_pred, y_true):
    # compare y_true to the index with the largest values across the column of y_pred
    # return the mean of the correct_prediction
    correct_prediction = tf.equal(tf.argmax(y_pred, 1), tf.cast(y_true, tf.int64))
    return tf.reduce_mean(tf.cast(correct_prediction, tf.float32), axis=-1)
```

```python
# run training for the given number of steps
for step, (batch_x, batch_y) in enumerate(train_data.take(training_steps), 1):
    # Run the optimization to update W and b values.
    run_optimization(batch_x, batch_y)
    
    # compute loss function and display stats after every 100 steps
    if step % display_step == 0:
        pred = neural_net(batch_x)
        loss = cross_entropy(pred, batch_y)
        acc = accuracy(pred, batch_y)
        print("Trained epochs: %i, Loss: %f, Accuracy: %f" % (step, loss, acc))
```

```python
# print accuracy of the test dataset
pred = neural_net(X_test)
print("Test Accuracy: %f" % accuracy(pred, y_test))
```

```python
# visualize the misclassified digit image data from the first 200 data in the test dataset
n_images = 200
test_images = X_test[:n_images]
test_labels = y_test[:n_images]
predictions = neural_net(test_images)

for i in range(n_images):
    model_prediction = np.argmax(predictions.numpy()[i])
    if (model_prediction != test_labels[i]):
        plt.imshow(np.reshape(test_images[i], [28, 28]), cmap="gray_r")
        plt.show()
        print("Original Labels: %i" % test_labels[i])
        print("Model prediction: %i" % model_prediction)
```
